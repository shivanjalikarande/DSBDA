```python
#Detecting Outliers
#1. Boxplot
import pandas as pd
import numpy as np
df = pd.read_csv("/home/pict/Documents/31442/Student_Performance.csv")
df
```

```python
col = ["math_score","reading_score","writing_score","placement_score"]
df.boxplot(col)
```

```python
#print outliers
print("Outlier in math score: ",np.where(df["math_score"]<60))
print("Outlier in reading score: ",np.where(df["reading_score"]<60))
print("Outlier in placement score: ",np.where(df["placement_score"]<70))

# array([ 3, 12])  -> Outliers present at index 3 and 12 in placement_score
```

```python
#2.Scatterplot
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(9,5))
ax.scatter(df["placement_score"],df["placement offer count"])
plt.show()
ax.set_xlabel("Placement Score")
ax.set_ylabel("Placement Offer Count")
```

```python
#print outliers
print(np.where((df['placement_score'] < 80) & (df['placement offer count'] > 1)))
```

```python
#3. Z-Score
from scipy import stats
z = np.abs(stats.zscore(df["reading_score"]))
print(z)
```

```python
#math_score contains NaN values -> Hence not executing zscore function
#Replace NaN by mean then execute 
mean_val = df["math_score"].mean()
df["math_score"].fillna(value = mean_val,inplace=True)
z = np.abs(stats.zscore(df["math_score"]))
print(z)
```

```python
#print outliers by setting threshold
threshold = 0.9
outliers = np.where(z < threshold)
outliers
```

```python
#4.IQR (Inter Quantile Range)

#sort column values
sorted_rscore = sorted(df["reading_score"])
sorted_rscore
```

```python
#calculates quantiles
q1 = np.percentile(sorted_rscore,25)
q3 = np.percentile(sorted_rscore,75)
IQR = q3 - q1

lower_bound = q1 - (1.5 * IQR)
upper_bound = q3 - (1.5 * IQR)
print("Lower Bound:",lower_bound,"\nUpper Bound:",upper_bound)
```

```python
#print outliers
reading_score_outliers = []
for i in sorted_rscore:
    if i < lower_bound or i > upper_bound:
        reading_score_outliers.append(i)
        
print("reading_score_outliers: ",reading_score_outliers)
```

```python
# Handling of outliers
#a. Trimming and removing outliers
new_df = df
for i in outliers:
    new_df.drop(i,inplace=True)
df
```

```python
#b.Quantile based flooring and capping
#outlier is capped at a certain value above the 90th percentile value or floored at a factor below the 10th percentile value

df_stud = df
ninetieth_percentile = np.percentile(df_stud['math_score'], 90)
new_math_score = np.where(df_stud["math_score"] > ninetieth_percentile, ninetieth_percentile,df_stud["math_score"])
print(new_math_score)
```

```python
#c.Mean Median Imputation
median = np.median(sorted_rscore)
print("Median in sorted reading_score: ",median)
refined_df = df
refined_df["reading_score"] = np.where(refined_df["reading_score"] > upper_bound,median,refined_df["reading_score"])
refined_df["reading_score"] = np.where(refined_df["reading_score"] < lower_bound,median,refined_df["reading_score"])
col = ["reading_score"]
refined_df.boxplot(col)
```
