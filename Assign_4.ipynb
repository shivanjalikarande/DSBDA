```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/home/pict/Downloads/titanic.csv")
print(dataset.head())
```

```python
import seaborn as sns
```

```python
#1. Distribution Plots
#a. Distplot
sns.distplot(x = dataset['Age'], bins = 10)

#bins = 10: This argument specifies the number of bins (bars) in the histogram. The data range will be divided into 10 equal-width intervals (bins), and each bin will show the count of data points that fall within that range.
```

```python
sns.distplot(dataset['Age'], bins = 10,kde=False)   #Remove the line representing kernel density estimation by kde = false
```

```python
#b.Join Plot - visually see the correlation or relationship between the two variables.

#b1 Scatter Plot -> A scatter plot shows individual data points as dots on the plot
sns.jointplot(x = dataset['Age'], y = dataset['Fare'], kind ='scatter')

#b2 Hexbin Plot -> Divides the space into hexagonal bins and colors them based on the density of points in each bin.
sns.jointplot(x = dataset['Age'], y = dataset['Fare'], kind ='hex')
```

```python
#c.Rug Plot - Used to draw small bars along the x-axis for each point in the dataset.
sns.rugplot(dataset["Fare"])
```

```python
#2. Categorical Plot

#a. Bar plot - used to display the mean value for each value in a categorical column(Sex), against a numeric column(Female)
sns.barplot(x = "Sex",y="Age",data=dataset)  #First parameter is categorical column , second parameter is numerical column
```

```python
#b. Count Plot - Displays the count of the categories in a specific column.
sns.countplot(x = 'Sex',data=dataset)
```

```python
#c.Box Plot -To display the distribution of the categorical data in the form of quartiles.
sns.boxplot(x ='Sex',y = 'Age',data=dataset)
```

```python
sns.boxplot(x='Sex',y='Age',data=dataset,hue='Survived')  # by adding another layer of distribution.
```

```python
#d.Violin Plot - Gives an idea of where the data is concentrated. Shows density of values
sns.violinplot(x='Sex',y='Age',data=dataset)
```

```python
sns.violinplot(x='Sex',y='Age',data=dataset,hue='Survived')
```

```python
#3. Advanced Plots
#a.Srip Plot - A strip plot displays each data point as a dot along the x-axis (or y-axis), with the position of the dot indicating the value of the variable being plotted.
sns.stripplot(x='Sex', y='Age', data=dataset, jitter=False)  
```

```python
sns.stripplot(x='Sex', y='Age', data=dataset, jitter=True)   # Jitter : Applied to spread out the points and avoid overlap.
```

```python
sns.stripplot(x='Sex', y='Age', data=dataset, jitter=True, hue='Survived')  #We can add additional categorical column
```

```python
#b.Swarm Plot - Combination of the strip and the violin plots, the points are adjusted in such a way that they don't overlap.
sns.swarmplot(x='Sex',y='Age',data=dataset)
```

```python
sns.swarmplot(x='Sex',y='Age',data=dataset,hue="Survived")
```

```python
#4. Matrix Plots

#a. Heat Maps - Used to plot correlation between numeric columns in the form of a matrix.
corr = dataset.corr() #To have useful information on both columns and row headers.
corr
sns.heatmap(corr)  #High correlation - White color, Low correlation - Black color
```

```python
sns.heatmap(corr,annot=True) #annot(annotations) - used to display the numerical values inside the heatmap cells.
```

```python
sns.heatmap(corr,cmap='plasma') #cmap - Changes the heatmap color scheme.
```

```python
sns.heatmap(corr,cmap='Greens')
```

```python
#b. Cluster map - Map basically uses Hierarchical Clustering to cluster the rows and columns of the matrix.
dataset_numeric = dataset.select_dtypes(include=["number"]).dropna()
dataset_numeric = dataset_numeric.set_index(dataset.columns[0])
print(dataset_numeric.head())
sns.clustermap(dataset_numeric,cmap='viridis')
```
