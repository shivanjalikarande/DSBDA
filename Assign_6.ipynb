```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset

df = pd.read_csv("/home/pict/31442_DSBDAL/Iris.csv")
```

```python
df.head(10)
```

```python
# 1. List the features and their types
print("\nFeatures and their types:")
print(df.dtypes)
```

```python
# 2. Create histograms for each feature
df.hist(bins=10, figsize=(12, 8))
plt.suptitle('Histograms of Iris Dataset Features')
plt.show()
```

```python
# 3. Create box plots for each feature
plt.figure(figsize=(12, 8))
for i, feature in enumerate(df.columns[:-1]):  # Exclude the 'species' column
    plt.subplot(2, 3, i+1)
    sns.boxplot(x='Species', y=feature, data=df)
    plt.title(f'Boxplot of {feature}')

plt.tight_layout()
plt.show()
```

```python
# Titanic Dataset
ndf = pd.read_csv("/home/pict/31442_DSBDAL/titanic.csv")
ndf.head(10)
```

```python
#Handle missing values for numerical columns
ndf['Age'].fillna(ndf['Age'].median())  
ndf['Fare'].fillna(ndf['Fare'].median())  
```

```python
# 1. List the features and their data types
print("\nFeatures and their types:")
print(ndf.dtypes)
```

```python
# 2. Create histograms for each feature
ndf.hist(bins=10, figsize=(12, 8))
plt.suptitle('Histograms of Titanic Dataset Features')
plt.show()
```

```python
# List of numerical columns to create box plots
numerical_features = ['Age', 'SibSp', 'Parch', 'Fare']
```

```python
# 3. Create box plots for each feature

plt.figure(figsize=(12,8))
for i, feature in enumerate(numerical_features):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(x='Survived', y=feature, data=ndf)  # 'Survived' is used on the x-axis
    plt.title(f'Boxplot of {feature}')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
