```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset

df = pd.read_csv("/home/pict/31442_DSBDAL/Assign_5/Iris.csv")
```

```python
df.head()
```

```python
# 1. List the features and their types
print("\nFeatures and their types:")
print(df.dtypes)
```

```python
# 2. Create histograms for each feature
df.hist(bins=10, figsize=(12, 8))
plt.suptitle('Histograms of Iris Dataset Features')
plt.show()
```

```python
# 3. Create box plots for each feature
plt.figure(figsize=(12, 8))
for i, feature in enumerate(df.columns[:-1]):  # Exclude the 'species' column
    plt.subplot(2, 3, i+1)
    sns.boxplot(x='Species', y=feature, data=df)
    plt.title(f'Boxplot of {feature}')

plt.tight_layout()
plt.show()
```
